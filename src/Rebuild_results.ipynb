```python
import os
import pandas as pd
import numpy as np
import csv
import json

try: 
    __file__
except:
    curr_dir = os.path.abspath('')
else:
    curr_dir = os.path.dirname(os.path.abspath(__file__))
    
app_root = curr_dir if os.path.basename(curr_dir) != "src" else os.path.dirname(curr_dir)

destdir = os.path.join(app_root,'data','manipulated_data')      # non-docker stay in repository


file = os.path.join(destdir,'Exogenous_better.csv')
halfgood_data = pd.read_csv(file)

halfgood_data
```

```python

halfgood_dict = {}

obj = halfgood_data['Target Location']

for key in obj.keys():
    value = halfgood_data['Target Location'][key]
    if isinstance(value, str):
        halfgood_dict[value] = { 'keymae': halfgood_data['Mean Absolute Error'][key], 'exogen': {} }

print(json.dumps(halfgood_dict, indent=4))
```

```python
obj2 = halfgood_data['Exogenous Location']

obj2_keys = obj2.keys()
dict_keys = list(halfgood_dict.keys())
current_dict_index = -1
collecting = False

i = 0
while i < len(obj2.keys()):
    value = obj2[obj2_keys[i]]
    if isinstance(value, str) and collecting:
        key = dict_keys[current_dict_index]
        value = value.strip(",").strip("'").strip(" ")
        halfgood_dict[key]['exogen'][value] = { 'exmae': halfgood_data['Mean Absolute Error'][i] }
    elif isinstance(value, str) and not collecting:
        collecting = True
        current_dict_index += 1
        i -= 1
    else:
        collecting = False
        
    i += 1

print(json.dumps(halfgood_dict, indent=4))
with open(os.path.join(destdir, "almostThere.json"), "w") as almost_there_file:
    almost_there_file.write( json.dumps(halfgood_dict, indent=4) )
```

Now that all the data is pulled out of a DataFrame and now in an equivalent Dictionary object.  Merge with AllMAE.json file

```python
with open(os.path.join(destdir, "manualThere.json"), "r") as manual_file:
    new_halfway_dict = json.loads(manual_file.read())

all_results_filename = os.path.join(destdir, 'allMAE.json')

with open(all_results_filename, "r") as all_results_file:
    all_results = json.loads(all_results_file.read())


target_location_list = list(new_halfway_dict.keys())

for i in range(0, len(target_location_list)):
    key = target_location_list[i]
    if key not in all_results:
        raise TypeError("Target Location Not Found ({0})".format(key))
        
    all_results[key]['keymae'] = new_halfway_dict[key]['keymae']
    
    exog_list = list(new_halfway_dict[key]['exogen'].keys())
    for j in range(0, len(exog_list)):
        exog_key = exog_list[j]
        if exog_key not in all_results[key]['exogen']:
            raise TypeError("ExogLocation in {0} Not Found ({1})".format(key, exog_key))
        
        all_results[key]['exogen'][exog_key]['exmae'] = new_halfway_dict[key]['exogen'][exog_key]['exmae']
        

# print(json.dumps(all_results, indent=4))
with open(all_results_filename, "w") as all_results_file:
    all_results_file.write( json.dumps(all_results, indent=4) )
```
